```python
from langchain.prompts import PromptTemplate

# Define a list of terrains
terrains = ["jungle", "desert", "mountainous", "coastal", "grassland", "arctic"]

# Loop through the terrains
for terrain in terrains:
    # Create a prompt 
    prompt_text = f"What would be a good name for a new national park with a {terrain} terrain?"
    print(f"\n{prompt_text}")
    
    # Create a HumanMessage schema object containing the prompt 
    messages = [HumanMessage(content=prompt_text)] 
    
    # Pass Messages to the chat model for inference
    response = chat_model.invoke(messages)
    
    # Print the generated names for the national park
    for index, name in enumerate(response.content.split("\n"), start=1):
        if name:
            print(f"{index}. {name}")
```

```python
# Define a list of terrains and wild animals
terrains = ["jungle", "desert", "mountainous"]
wild_animals = ["tigers", "lions", "bears"]

# Create a prompt question
prompt_question = "What would be a good name for a new national park with a {terrain} terrain and {animal}?"

# Print the prompt question
print(f"{prompt_question}\n")

# Loop through terrains and wild animals
for terrain in terrains:
    for animal in wild_animals:
        # Create a prompt 
        prompt_text = prompt_question.format(terrain=terrain, animal=animal)
        
        # Create a HumanMessage schema object containing the prompt 
        messages = [HumanMessage(content=prompt_text)] 
        
        # Pass Messages to the chat model for inference
        response = chat_model.invoke(messages)
        
        # Print the generated names
        for name in response.content.split("\n"):
            if name:
                print(f"Terrain: {terrain}, Wild Animal: {animal}, Generated Name: {name}\n")
```

```python
from langchain.prompts import ChatPromptTemplate
from langchain.schema.messages import AIMessage, HumanMessage

# Create a chat prompt template
chat_template = ChatPromptTemplate.from_messages([
    ("system", "If given name of the terrain and wild animal, you will generate the name of the national park."),
    ("human", "{text}"),
])

# Format messages using the chat template
llm = ChatOpenAI()
llm(chat_template.format_messages(text="bears on mountain"))
```

```python
from langchain.llms import OpenAI
from langchain.prompts import ChatPromptTemplate

park_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a avid traveller of nature. You like to give suggestions of national park for travel."),
        ("system", "You always make up a new name of national park and suggest."),
        ("human", "Hello, how are you doing?"),
        ("ai", "I'm doing well, thanks! Give me suggestion of animal or terrain."),
        ("human", "{animal} {terrain}"),
        ("human", "{text}"),
    ]
)

messages = park_template.format_messages(animal="bears", terrain="mountains", text="")
response = llm(messages)

# Format output
output = f"""
Travel Advisor Response:

{response.content}
"""

print(output)
```

```python
# User requests different animal and terrain
messages = park_template.format_messages(
  animal="penguins", 
  terrain="deserts",
  text=""  
)

response = llm(messages)

# Format output
output = f"""
Travel Advisor Response:

{response.content}
"""

print(output)
```
